## **3. Análisis Exploratorio de Datos (EDA) y Visualización de Resultados**

Ahora, se procede a realizar un **Análisis Exploratorio de Datos (EDA)**. Para ello, se utilizó una herramienta que permite generar un reporte EDA automático, facilitando la exploración y comprensión de los datos de manera rápida y eficiente.

Para utilizar esta herramienta, instalaremos la librería **`dlookr`**


```r
install.packages("dlookr")
library("dlookr")
```

**Dlookr tiene una función llamada `eda_web_report()` que permite generar un reporte automático del análisis exploratorio de datos (EDA). A continuación se detallan los parámetros utilizados en esta función:**

- `target`: Representa la variable de salida a estudiar. En este caso, hemos seleccionado **GradeClass** como la variable objetivo, ya que clasifica el rendimiento académico de los estudiantes según su GPA.
  
- `output_file`: Especifica el nombre del archivo del reporte cuando se descargue. En este ejemplo, el archivo se llamará **reporte_EDA_con_R.html**.

- `output_dir`: Indica el directorio de salida donde se guardará el reporte generado. Se utiliza la función `getwd()` para obtener el directorio de trabajo actual.

- `titel`: Define el título del reporte. En este caso, es **"Reporte EDA-R"**.

- `subtitle`: Especifica el subtítulo del reporte. En este ejemplo, se ha utilizado **"Actividad con R"**.

- `theme`: Permite personalizar el tema del reporte. En este caso, se ha elegido el tema **"blue"**.

Este conjunto de parámetros genera un reporte interactivo que facilita la comprensión y visualización de los datos, lo cual es útil para un análisis más profundo y detallado.


Elegí como variable objetivo la columna GradeClass, ya que esta columna clasifica las calificaciones de los estudiantes en función de su GPA, lo que la convierte en una medida clave del rendimiento académico. Esta variable es influenciada por factores como el tiempo de estudio, la participación en actividades extracurriculares, el apoyo de los padres, entre otros. Por lo tanto, GradeClass representa una forma efectiva de categorizar el rendimiento académico de los estudiantes y entender cómo los diferentes factores impactan en su desempeño.

```r
eda_web_report(df,
               target="GradeClass",
               output_file="reporte_EDA_con_R.html",
               output_dir= getwd(),
                titel= "Reporte EDA-R",
                subtitle= "Actividad con R",
                theme="blue"
              )
```

Según el reporte, no se encuentran datos nulos, pero verificaremos si hay filas o datos duplicados.

```r
# Verificar si hay filas duplicadas
duplicated_rows <- duplicated(df)
sum(duplicated_rows)  # Cuenta cuántas filas están duplicadas
```

Ahora verificaremos únicamente si hay duplicados en la columna `StudentID`, ya que es el único caso en el que tendría sentido eliminar los registros duplicados, ya que los demás son datos numéricos que pueden repetirse entre diferentes estudiantes.

```r
# Verificar duplicados solo por 'StudentID'
duplicated_ids <- duplicated(df$StudentID)
sum(duplicated_ids)  # Cuenta cuántos duplicados hay en 'StudentID'
```

# **Conclusiones Generales**

1. **Influencia significativa del GPA en GradeClass:**  
   Se observó que el **GPA** tiene una fuerte correlación negativa con **GradeClass**, indicando que a medida que el rendimiento académico mejora (mayor **GPA**), las calificaciones de los estudiantes también mejoran. Esto refuerza la idea de que el **GPA** es un predictor clave para clasificar el rendimiento académico y debe ser considerado como una variable principal en cualquier modelo predictivo.

2. **Importancia de la asistencia a clases:**  
   La correlación positiva significativa entre **Ausencias** y **GradeClass** demuestra que los estudiantes con mayor cantidad de ausencias tienden a obtener calificaciones más bajas. Esto subraya la necesidad de incluir la variable **Ausencias** como un factor crítico en los modelos de predicción, ya que su impacto en el rendimiento académico es considerable.

3. **Relevancia limitada del tiempo de estudio semanal:**  
   Aunque la relación entre **StudyTimeWeekly** y **GradeClass** es débil, no debe descartarse por completo. Es posible que esta variable tenga un impacto más significativo cuando se analiza en combinación con otras características, como el **GPA** o las **Ausencias**.

4. **Impacto indirecto del apoyo parental:**  
   La débil correlación negativa entre **ParentalSupport** y **GradeClass** sugiere que, aunque el apoyo de los padres no tiene un impacto directo fuerte en las calificaciones, podría influir indirectamente en otras variables, como los hábitos de estudio o la asistencia.
